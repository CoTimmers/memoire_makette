## Newton–Euler Equations (3 equations, 3 unknowns)

Contact model (sliding friction):

$$
f_A = \begin{bmatrix} -\mu\,f_{yA} \\ f_{yA} \end{bmatrix}, \qquad
f_B = f_{Bn}\,\hat{n}_B - \mu\,f_{Bn}\,\hat{t}_B
$$

The 3 Newton–Euler equations:

$$
\begin{cases}
f_A[0] + f_B[0] + f_x = m\,a_x \\[6pt]
f_A[1] + f_B[1] + f_y = m\,a_y \\[6pt]
r_{OA} \times f_A + r_{OB} \times f_B = I_A\,\ddot\delta
\end{cases}
$$

which gives the linear system in $(\ddot\ell,\, f_{yA},\, f_{Bn})$:

$$
\begin{bmatrix}
m\,c_x           & \mu          & -f_{Bdir}[0] \\
m\,c_y           & -1           & -f_{Bdir}[1] \\
I_A\,\delta_\ell & 0            & -(r_{OB} \times f_{Bdir})
\end{bmatrix}
\begin{bmatrix}
\ddot\ell \\ f_{yA} \\ f_{Bn}
\end{bmatrix}
=
\begin{bmatrix}
f_x - m\,a_x^{nl} \\
f_y - m\,a_y^{nl} \\
r_{OA} \times \begin{bmatrix}f_x\\f_y\end{bmatrix} - I_A\,\ddot\delta^{nl}
\end{bmatrix}
$$

with $f_{Bdir} = \hat{n}_B - \mu\,\hat{t}_B$ the contact direction at B, and:

$$
c_x = 1 + (R'(\delta)\,r_{AO})[0]\cdot\delta_\ell, \qquad c_y = (R'(\delta)\,r_{AO})[1]\cdot\delta_\ell
$$

# Modèle et loi de commande

## 1. Erreur de position

La position cible du chariot, obtenue grâce à la calibration de la caméra, est notée $x_{\text{target}}$.  
La position réelle du chariot, mesurée par les encodeurs, est notée $x(t)$.

L'erreur de position est définie par :

$$
e(t) = x_{\text{target}} - x(t)
$$

---

## 2. Modèle cinématique simplifié du chariot

Dans une première approximation, le mouvement horizontal du chariot peut être modélisé par :

$$
\dot{x}_{\text{crane}}(t) = v_{\text{cmd}}(t)
$$

où :

- $x(t)$ est la position du chariot,
- $v(t)$ est la vitesse commandée.

---

## 3. Loi de vitesse par zones

La stratégie choisie repose sur une consigne de vitesse discrète, définie en fonction de l'erreur de position :

$$
v_d(t)=
\begin{cases}
\operatorname{sgn}(e(t))\,v_{\text{fast}} & \text{if } |e(t)| > d_{\text{near}} \\
\operatorname{sgn}(e(t))\,v_{\text{slow}} & \text{if } e_{\text{stop}} < |e(t)| \le d_{\text{near}} \\
0 & \text{if } |e(t)| \le e_{\text{stop}}
\end{cases}
$$

où :

- $v_{\text{fast}}$ est la vitesse rapide,
- $v_{\text{slow}}$ est la vitesse lente,
- $d_{\text{near}}$ est le seuil de la zone proche de la cible,
- $e_{\text{stop}}$ est la bande d'arrêt,
- $\operatorname{sgn}(e)$ donne le sens du mouvement.

---

## 4. Limitation de l'accélération

La vitesse effectivement envoyée au moteur, notée $v_{\text{cmd}}(t)$, ne peut pas varier instantanément.  
On impose donc une contrainte sur son évolution :

$$
\dot{v}_{\text{cmd}}(t) = a(t)
$$

avec la contrainte :

$$
|a(t)| = |\dot{v}_{\text{cmd}}(t)| \le a_{\max}
$$

Autrement dit :

$$
|\dot{v}_{\text{cmd}}(t)| \le a_{\max}
$$

Cette limitation permet de réduire les variations brutales de vitesse et donc de limiter l'excitation du balancement de la charge suspendue.

---

## 5. Modèle dynamique de la charge suspendue

La charge suspendue est assimilée à un pendule de longueur $L$.  
Le modèle dynamique exact est :

$$
\ddot{\theta}(t) + \frac{g}{L}\sin\big(\theta(t)\big) = \frac{\ddot{x}(t)}{L}
$$

où :

- $\theta(t)$ est l'angle de balancement,
- $g$ est l'accélération gravitationnelle,
- $L$ est la longueur du câble,
- $\ddot{x}(t)$ est l'accélération horizontale du chariot.

---

## 6. Modèle linéarisé pour petits angles

Dans le domaine de fonctionnement visé, le débattement reste faible.  
On peut donc utiliser l'approximation des petits angles :

$$
\sin(\theta) \approx \theta
$$

Le modèle devient alors :

$$
\ddot{\theta}(t) + \frac{g}{L}\theta(t) = \frac{\ddot{x}(t)}{L}
$$

Ce modèle montre que l'accélération horizontale du chariot agit directement comme une excitation du balancement.

---

## 7. Relation simplifiée entre angle et accélération

Dans une situation quasi statique ou pour une interprétation de premier ordre, on peut négliger temporairement le terme $\ddot{\theta}(t)$, ce qui donne :

$$
\frac{g}{L}\theta(t) \approx \frac{\ddot{x}(t)}{L}
$$

d'où :

$$
\theta(t) \approx \frac{\ddot{x}(t)}{g}
$$

Cette relation simplifiée permet d'estimer l'effet de l'accélération du chariot sur le débattement angulaire de la charge.

---

## 8. Distances minimales de ralentissement

Pour garantir que la vitesse lente est atteinte avant l'entrée dans la zone proche, il faut anticiper le freinage.

En utilisant la relation cinématique :

$$
v_f^2 = v_i^2 + 2 a \Delta x
$$

on obtient la distance minimale nécessaire pour ralentir de $v_{\text{fast}}$ à $v_{\text{slow}}$ avec une décélération maximale $a_{\max}$ :

$$
d_{fs} = \frac{v_{\text{fast}}^2 - v_{\text{slow}}^2}{2a_{\max}}
$$

où :

- $d_{fs}$ est la distance minimale de transition entre la vitesse rapide et la vitesse lente.

De même, la distance minimale nécessaire pour ralentir de $v_{\text{slow}}$ à l'arrêt vaut :

$$
d_{s0} = \frac{v_{\text{slow}}^2}{2a_{\max}}
$$

où :

- $d_{s0}$ est la distance minimale pour passer de la vitesse lente à la vitesse nulle.

---

## 9. Condition d'anticipation du ralentissement

Si la distance restante à la cible est notée :

$$
r(t) = |e(t)|
$$

alors :

- le ralentissement de $v_{\text{fast}}$ vers $v_{\text{slow}}$ doit commencer avant l'entrée dans la zone proche,
- le ralentissement final de $v_{\text{slow}}$ vers $0$ doit commencer avant la cible.

Ainsi, pour entrer dans la zone proche déjà à vitesse lente, il faut prévoir une distance d'anticipation égale à :

$$
d_{\text{near}} + d_{fs}
$$

et pour s'arrêter correctement à la cible, il faut prévoir une distance minimale :

$$
d_{s0}
$$

---

## 10. Résumé de la stratégie

La commande repose donc sur les étapes suivantes :

1. calcul de l'erreur :

$$
e(t)=x_{\text{target}}-x(t)
$$

2. choix d'une vitesse désirée discrète :

$$
v_d(t)\in\{ -v_{\text{fast}}, -v_{\text{slow}}, 0, v_{\text{slow}}, v_{\text{fast}} \}
$$

3. application d'une contrainte d'accélération :

$$
|\dot{v}_{\text{cmd}}(t)| \le a_{\max}
$$

4. prise en compte du modèle pendulaire linéarisé :

$$
\ddot{\theta}(t) + \frac{g}{L}\theta(t) = \frac{\ddot{x}(t)}{L}
$$

5. anticipation des ralentissements à l'aide des distances :

$$
d_{fs} = \frac{v_{\text{fast}}^2 - v_{\text{slow}}^2}{2a_{\max}}
$$

$$
d_{s0} = \frac{v_{\text{slow}}^2}{2a_{\max}}
$$

\end{bmatrix}
$$

où :

- $x(t)$ : position du chariot  
- $v(t)$ : vitesse du chariot  
- $a(t) = \dot{v}(t)$ : accélération commandée  

avec la contrainte :

$$
|a(t)| \le a_{\max}
$$

---

## 2. États de la charge suspendue (pendule)

La charge est modélisée comme un pendule de longueur $L$.

On définit les états :

$$
\mathbf{x}_p(t) =
\begin{bmatrix}
\theta(t) \\
\dot{\theta}(t)
\end{bmatrix}
$$

où :

- $\theta(t)$ : angle de déviation  
- $\dot{\theta}(t)$ : vitesse angulaire  

---

## 3. Modèle dynamique du pendule

### Modèle exact

$$
\ddot{\theta}(t) + \frac{g}{L}\sin(\theta(t)) = \frac{\ddot{x}(t)}{L}
$$

---

### Modèle linéarisé (petits angles)

$$
\ddot{\theta}(t) + \frac{g}{L}\theta(t) = \frac{\ddot{x}(t)}{L}
$$

---

## 4. Mise sous forme d'état

On peut écrire le système du pendule sous forme d'état :

$$
\dot{\mathbf{x}}_p(t) =
\begin{bmatrix}
\dot{\theta}(t) \\
\ddot{\theta}(t)
\end{bmatrix}
=
\begin{bmatrix}
\dot{\theta}(t) \\
-\frac{g}{L}\theta(t) + \frac{1}{L}\ddot{x}(t)
\end{bmatrix}
$$

---

## 5. Système complet

Le système global combine le chariot et la charge :

$$
\mathbf{x}(t) =
\begin{bmatrix}
x(t) \\
\dot{x}(t) \\
\theta(t) \\
\dot{\theta}(t)
\end{bmatrix} 



$$

avec la dynamique :

$$
\dot{\mathbf{x}}(t) =
\begin{bmatrix}
v(t) \\
a(t) \\
\dot{\theta}(t) \\
-\frac{g}{L}\theta(t) + \frac{1}{L}a(t)
\end{bmatrix}
$$

---

## 6. Interprétation

- Le chariot est directement commandé par l'accélération $a(t)$  
- La charge est influencée indirectement par cette accélération  
- Le terme $\frac{1}{L}a(t)$ agit comme une excitation du pendule  

---

## 7. Lien avec la stratégie de commande

La contrainte :

$$
|a(t)| \le a_{\max}
$$

permet de limiter :

$$
\theta(t) \approx \frac{\ddot{x}(t)}{g}
$$

et donc de contrôler le débattement de la charge.

## Génération de la consigne de vitesse

La consigne de vitesse est définie comme une fonction continue de l'erreur de position :

$$
v_{\text{ref}}(t) = f\big(e_x(t)\big)
$$

avec :

$$
e_x(t) = x_{\text{target}} - x(t)
$$

Cette consigne n'est pas modifiée brutalement, mais respecte des contraintes dynamiques afin de limiter l'excitation du pendule :

$$
\left| \dot{v}_{\text{ref}}(t) \right| \le a_{\max}
\quad \text{and} \quad
\left| \ddot{v}_{\text{ref}}(t) \right| \le j_{\max}
$$

où :

- $a_{\max}$ est l'accélération maximale admissible,
- $j_{\max}$ est le jerk maximal (variation de l'accélération).



## State transition condition

A transition to the next state is triggered by a completion flag based on tolerance conditions.

The position error must satisfy:

$$
|x_{\text{target}} - x(t)| \le \varepsilon_x
$$

and the velocity must be sufficiently small:

$$
|v(t)| \le \varepsilon_v
$$

Optionally, additional conditions can be included to account for the pendulum dynamics:

$$
|\theta(t)| \le \varepsilon_\theta
\quad \text{and} \quad
|\dot{\theta}(t)| \le \varepsilon_{\dot{\theta}}
$$

The transition flag is therefore defined as:

$$
\text{flag}_{\text{target reached}} =
\begin{cases}
1 & \text{if all conditions are satisfied} \\
0 & \text{otherwise}
\end{cases}
$$

# System Model and Key Properties

## Dynamic model

The suspended container can be modeled as a damped second-order system:

$$
\ddot{\theta} + 2\zeta\omega_n \dot{\theta} + \omega_n^2 \theta = -\frac{1}{l} a(t)
$$

where:

- $\theta(t)$ is the angular displacement  
- $a(t)$ is the acceleration input  
- $\omega_n = \sqrt{\frac{g}{l}}$ is the natural frequency  
- $\zeta$ is the damping ratio  

---

## Transfer function

Taking the Laplace transform with zero initial conditions:

$$
\frac{\Theta(s)}{A(s)} = \frac{-1/l}{s^2 + 2\zeta\omega_n s + \omega_n^2}
$$

---

## Poles and system behavior

The poles are given by the roots of the denominator:

$$
s^2 + 2\zeta\omega_n s + \omega_n^2 = 0
$$

The discriminant is:

$$
\Delta = (2\zeta\omega_n)^2 - 4\omega_n^2
$$

If $\Delta < 0$:
 
 , the poles are complex conjugates  


- This corresponds to an **underdamped system**  
- The response exhibits **oscillations with exponential decay**  

Therefore:

A negative discriminant ($\Delta < 0$) implies a damped oscillatory behavior.

---

## Step input

A step acceleration input is defined as:

$$
a(t) = A_0 \, u(t)
$$

In Laplace domain:

$$
A(s) = \frac{A_0}{s}
$$

---

## Steady-state angle

Under constant acceleration, the system tends toward:

$$
\theta_{eq} = -\frac{A_0}{g}
$$

---

## Peak response with overshoot

For an underdamped system, the maximum angle reached is:

$$
\theta_{peak} = \theta_{eq} (1 + M_p)
$$

where the overshoot is:

$$
M_p = e^{-\frac{\zeta \pi}{\sqrt{1-\zeta^2}}}
$$

---

## Interpretation

- A step input excites the natural oscillatory mode  
- The system oscillates around the equilibrium angle  
- The amplitude of oscillation depends on the damping ratio  
- The peak angle can exceed the steady-state value due to overshoot  

# Derivation of the Standard Second-Order Model

## 1. Starting point: dynamic equation

The system is modeled as a damped pendulum with an acceleration input:

$$
J\ddot{\theta} + c\dot{\theta} + mgl \sin(\theta) = -ml\,a(t)
$$

where:

- $J$ is the moment of inertia  
- $c$ is the damping coefficient  
- $m$ is the mass  
- $l$ is the cable length  

---

## 2. Linearization (small-angle approximation)

Assuming small angles:

$$
\sin(\theta) \approx \theta
$$

The equation becomes:

$$
J\ddot{\theta} + c\dot{\theta} + mgl\,\theta = -ml\,a(t)
$$

---

## 3. Substitute inertia

For a simple pendulum:

$$
J = ml^2
$$

Substitute into the equation:

$$
ml^2 \ddot{\theta} + c\dot{\theta} + mgl\,\theta = -ml\,a(t)
$$

---

## 4. Normalize the equation

Divide both sides by $ml^2$:

$$
\ddot{\theta} + \frac{c}{ml^2}\dot{\theta} + \frac{g}{l}\theta = -\frac{1}{l}a(t)
$$

---

## 5. Standard second-order form

We rewrite the equation as:

$$
\ddot{\theta} + 2\zeta \omega_n \dot{\theta} + \omega_n^2 \theta = -\frac{1}{l} a(t)
$$

---

## 6. Identification of parameters

By comparison:

### Natural frequency

$$
\omega_n = \sqrt{\frac{g}{l}}
$$

### Damping ratio

$$
2\zeta\omega_n = \frac{c}{ml^2}
$$

Thus:

$$
\zeta = \frac{c}{2ml^2\omega_n}
$$

---

## 7. Final model

$$
\ddot{\theta} + 2\zeta \omega_n \dot{\theta} + \omega_n^2 \theta = -\frac{1}{l} a(t)
$$

---

## Interpretation

- $\omega_n$ defines the natural oscillation frequency  
- $\zeta$ characterizes the damping level  
- The right-hand side represents the excitation due to base acceleration  

# Transfer Function Derivation in the Laplace Domain

Starting from the standard second-order model:

$$
\ddot{\theta} + 2\zeta \omega_n \dot{\theta} + \omega_n^2 \theta = -\frac{1}{l} a(t)
$$

Assuming zero initial conditions, the Laplace transform gives:

$$
s^2 \Theta(s) + 2\zeta \omega_n s \Theta(s) + \omega_n^2 \Theta(s) = -\frac{1}{l} A(s)
$$

Factoring out $\Theta(s)$:

$$
\Theta(s)\left(s^2 + 2\zeta \omega_n s + \omega_n^2\right) = -\frac{1}{l} A(s)
$$

Therefore:

$$
\Theta(s) = \frac{-\frac{1}{l}A(s)}{s^2 + 2\zeta \omega_n s + \omega_n^2}
$$

---

## Transfer function

The transfer function is:

$$
\frac{\Theta(s)}{A(s)} = \frac{-1/l}{s^2 + 2\zeta \omega_n s + \omega_n^2}
$$

---

## Step input

For a step acceleration input of amplitude $A_0$:

$$
a(t) = A_0 u(t)
$$

its Laplace transform is:

$$
A(s) = \frac{A_0}{s}
$$

Substituting into the previous equation:

$$
\Theta(s) = \frac{-\frac{1}{l}\frac{A_0}{s}}{s^2 + 2\zeta \omega_n s + \omega_n^2}
$$

which can be written as:

$$
\Theta(s) = \frac{-A_0/l}{s\left(s^2 + 2\zeta \omega_n s + \omega_n^2\right)}
$$


## 2. What does this expression tell us?

The denominator contains:

- a pole at $s = 0$ coming from the step input  
- the two natural poles of the system:

$$
s_{1,2} = -\zeta \omega_n \pm j \omega_n \sqrt{1 - \zeta^2}
$$

if $\zeta < 1$.

---

Therefore, the time response contains:

- a constant component  
- a damped oscillatory component  

👉 Therefore, a step input excites vibrations in $\theta$.
