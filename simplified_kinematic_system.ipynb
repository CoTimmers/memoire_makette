### Kinematics and Dynamics of the System (ℓ, ψ, δ)

![fbd.jpeg](../memoire_makette/pivoting_wall.png)

---

#### 1. Reference Frames

We define:

- $\ell(t)$: horizontal translation along the world $x$-axis.  
- $\psi(t)$: imposed wall angle.  
- $\delta(t)$: orientation of the box.  

Constant vector corner → COM:

$$
r = \begin{bmatrix} a/2 \\ b/2 \end{bmatrix}.
$$

Contact points (body frame, origin at COM):

$$
r_A^{\text{local}} =
\begin{bmatrix}
-a/2 \\
-b/2
\end{bmatrix},
\qquad
r_B^{\text{local}} =
\begin{bmatrix}
-a/2 \\
+b/2
\end{bmatrix}.
$$

---

#### 2. Rotation Matrix $R(\delta)$ and Derivative $R'(\delta)$


$$
R(\delta)=
\begin{bmatrix}
\cos\delta & -\sin\delta \\
\sin\delta & \cos\delta
\end{bmatrix}.
$$

$$
R'(\delta)=
\frac{\partial R}{\partial \delta}=
\begin{bmatrix}
-\sin\delta & -\cos\delta \\
\cos\delta & -\sin\delta
\end{bmatrix}.
$$

---

#### 3. Position of the Center of Mass $(x_c,y_c)$

Translation along $x$:

$$
\ell_{\text{vec}} =
\begin{bmatrix}
\ell \\
0
\end{bmatrix}.
$$

Position of the COM in the world frame:

$$
p =
\begin{bmatrix}
x_c \\
y_c
\end{bmatrix}
=
\ell_{\text{vec}} + R(\delta)\,r.
$$

---

#### 4. Geometric Relation $\delta(\ell,\psi)$

Delta is found with the sinus law and with the sum of the angles of a triangle is equal to 180°. 

$$
\delta(\ell,\psi)
=
\frac{\pi}{2}
-
\psi
-
\arcsin\!\left(
\frac{\ell}{b}\cos\psi
\right)
$$

Define:

$$
u(\ell,\psi)
=
\frac{\ell}{b}\cos\psi.
$$

---

#### 5. Derivatives of $\delta$

$$
\frac{d}{du}\arcsin(u)
=
\frac{1}{\sqrt{1-u^2}}
$$

Define:

$$
k =
\frac{1}{\sqrt{1-u^2}}
$$

Second derivative term:

$$
\frac{dk}{du}
=
u(1-u^2)^{-3/2}
=
u\,k^3
$$

with

$$
k^3 =
\frac{1}{(1-u^2)^{3/2}}.
$$

Condition:

$$
|u| < 1
\quad \Longrightarrow \quad
\left|\frac{\ell}{b}\cos\psi\right| < 1.
$$

---

##### Derivatives of $u$

$$
u_\ell = \frac{1}{b}\cos\psi
$$

$$
u_\psi = -\frac{\ell}{b}\sin\psi
$$

$$
u_{\ell\ell}=0
$$

$$
u_{\ell\psi}=-\frac{1}{b}\sin\psi
$$

$$
u_{\psi\psi}=-\frac{\ell}{b}\cos\psi
$$

---

##### First derivatives of $\delta$

$$
\boxed{
\delta_\ell
=
-
k\,\frac{1}{b}\cos\psi
}
$$

$$
\boxed{
\delta_\psi
=
-1
+
k\,\frac{\ell}{b}\sin\psi
}
$$

---

##### Second derivatives of $\delta$

$$
\boxed{
\delta_{\ell\ell}
=
-
u\,k^3\,u_\ell^2
}
$$

$$
\boxed{
\delta_{\ell\psi}
=
-
\left(
k\,u_{\ell\psi}
+
u\,k^3\,u_\ell u_\psi
\right)
}
$$

$$
\boxed{
\delta_{\psi\psi}
=
-
\left(
k\,u_{\psi\psi}
+
u\,k^3\,u_\psi^2
\right)
}
$$

---

$$
\boxed{
\dot\delta
=
\delta_\ell\,\dot\ell
+
\delta_\psi\,\dot\psi
}
$$


$$
\boxed{
\ddot\delta
=
\delta_{\ell\ell}\,\dot\ell^2
+
2\,\delta_{\ell\psi}\,\dot\ell\dot\psi
+
\delta_{\psi\psi}\,\dot\psi^2
+
\delta_\ell\,\ddot\ell
+
\delta_\psi\,\ddot\psi
}
$$


---

#### 7. Velocity & acceleration

Differentiating $p(t)$:

$$
\dot p =
\begin{bmatrix}
\dot\ell \\
0
\end{bmatrix}
+
R'(\delta)\,r\,\dot\delta.
$$


Differentiating $p\dot(t)$:

$$
\boxed{
\ddot p =
\begin{bmatrix}
\ddot\ell \\
0
\end{bmatrix}
+
\big(R'(\delta)r\big)\,\ddot\delta
-
\big(R(\delta)r\big)\,(\dot\delta)^2
}
$$

Define:

$$
a_x = (\ddot p)_x,
\qquad
a_y = (\ddot p)_y.
$$

---

#### 8. Newton–Euler Equations (3 Equations)

$$
\left\{
\begin{aligned}
-\mu f_{xA}\,\mathrm{sign}(\dot\ell)
-\mu f_{xB}\,\mathrm{sign}(\dot\ell)
+ f_x
&= m\,a_x \\[6pt]

f_{yA} + f_{yB} + f_y
&= m\,a_y \\[6pt]

(r_{OA}\times f_A) + (r_{OB}\times f_B)
&= I\,\ddot\delta
\end{aligned}
\right.
$$

with

$$
r_{OA}=R(\delta)\,r_A^{\text{local}},
\qquad
r_{OB}=R(\delta)\,r_B^{\text{local}}.
$$



$$
f_{xA} = -\mu\,f_{yA}\,\mathrm{sign}(\dot\ell),
\qquad
f_{xB} = -\mu\,f_{yB}\,\mathrm{sign}(\dot\ell).
$$


$$
f_{yA} \ge 0,
\qquad
f_{yB} \ge 0.
$$


---

#### 9. Solution Procedure (Time Step $t_k$)

1. Known: $(\ell,\dot\ell)$ from integration, and imposed $(\psi,\dot\psi,\ddot\psi)$.

$$
\dot\ell_{k+1}=\dot\ell_k+\ddot\ell_k\Delta t,
\qquad
\ell_{k+1}=\ell_k+\dot\ell_{k+1}\Delta t.
$$

2. Compute $\delta(\ell,\psi)$, then $\dot\delta$ and $\ddot\delta$.
3. Compute $(a_x,a_y)$.
4. Solve the 3 Newton–Euler equations for $(\ddot\ell,f_{yA},f_{yB})$.
