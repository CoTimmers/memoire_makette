## Newton–Euler Equations (3 equations, 3 unknowns)

Contact model (sliding friction):

$$
f_A = \begin{bmatrix} -\mu\,f_{yA} \\ f_{yA} \end{bmatrix}, \qquad
f_B = f_{Bn}\,\hat{n}_B - \mu\,f_{Bn}\,\hat{t}_B
$$

The 3 Newton–Euler equations:

$$
\begin{cases}
f_A[0] + f_B[0] + f_x = m\,a_x \\[6pt]
f_A[1] + f_B[1] + f_y = m\,a_y \\[6pt]
r_{OA} \times f_A + r_{OB} \times f_B = I_A\,\ddot\delta
\end{cases}
$$

which gives the linear system in $(\ddot\ell,\, f_{yA},\, f_{Bn})$:

$$
\begin{bmatrix}
m\,c_x           & \mu          & -f_{Bdir}[0] \\
m\,c_y           & -1           & -f_{Bdir}[1] \\
I_A\,\delta_\ell & 0            & -(r_{OB} \times f_{Bdir})
\end{bmatrix}
\begin{bmatrix}
\ddot\ell \\ f_{yA} \\ f_{Bn}
\end{bmatrix}
=
\begin{bmatrix}
f_x - m\,a_x^{nl} \\
f_y - m\,a_y^{nl} \\
r_{OA} \times \begin{bmatrix}f_x\\f_y\end{bmatrix} - I_A\,\ddot\delta^{nl}
\end{bmatrix}
$$

with $f_{Bdir} = \hat{n}_B - \mu\,\hat{t}_B$ the contact direction at B, and:

$$
c_x = 1 + (R'(\delta)\,r_{AO})[0]\cdot\delta_\ell, \qquad c_y = (R'(\delta)\,r_{AO})[1]\cdot\delta_\ell
$$

## Newton–Euler Equations (3 equations, 3 unknowns)

Given $(\ell, \dot\ell, \psi, \dot\psi, \ddot\psi)$, compute $a_x$, $a_y$ and $\ddot\delta$:

$$
a_x = \ddot\ell + (R'(\delta)\,r_{AO})_x\,\ddot\delta - (R(\delta)\,r_{AO})_x\,\dot\delta^2
$$

$$
a_y = (R'(\delta)\,r_{AO})_y\,\ddot\delta - (R(\delta)\,r_{AO})_y\,\dot\delta^2
$$

$$
\ddot\delta = \delta_{\ell\ell}\,\dot\ell^2 + 2\,\delta_{\ell\psi}\,\dot\ell\dot\psi + \delta_{\psi\psi}\,\dot\psi^2 + \delta_\psi\,\ddot\psi + \delta_\ell\,\ddot\ell
$$

Contact model (sliding friction):

$$
f_{xA} = -\mu\,f_{yA}\,\mathrm{sign}(\dot\ell), \qquad f_{xB} = -\mu\,f_{yB}\,\mathrm{sign}(\dot\ell)
$$

The 3 Newton–Euler equations:

$$
\begin{cases}
f_{xA} + f_{xB} + f_x = m\,a_x \\[6pt]
f_{yA} + f_{yB} + f_y = m\,a_y \\[6pt]
r_{OA} \times f_A + r_{OB} \times f_B = I_A\,\ddot\delta
\end{cases}
$$

Substituting the contact model and expanding $a_x$, $a_y$, $\ddot\delta$ in $\ddot\ell$, the system becomes linear in $(\ddot\ell,\, f_{yA},\, f_{yB})$:

$$
\underbrace{
\begin{bmatrix}
m\,c_x        & \mu  & \mu  \\
m\,c_y        & -1   & -1   \\
I_A\,\delta_\ell & r_{OA,x} & -r_{OB,x}
\end{bmatrix}}_{M}
\begin{bmatrix}
\ddot\ell \\ f_{yA} \\ f_{yB}
\end{bmatrix}
=
\underbrace{
\begin{bmatrix}
f_x - m\,a_x^{nl} \\
f_y - m\,a_y^{nl} \\
-I_A\,\ddot\delta^{nl}
\end{bmatrix}}_{\text{rhs}}
$$

where the nonlinear terms (known at each timestep) are:

$$
a_x^{nl} = (R'(\delta)\,r_{AO})_x\,\ddot\delta^{nl} - (R(\delta)\,r_{AO})_x\,\dot\delta^2
$$

$$
a_y^{nl} = (R'(\delta)\,r_{AO})_y\,\ddot\delta^{nl} - (R(\delta)\,r_{AO})_y\,\dot\delta^2
$$

$$
\ddot\delta^{nl} = \delta_{\ell\ell}\,\dot\ell^2 + 2\,\delta_{\ell\psi}\,\dot\ell\dot\psi + \delta_{\psi\psi}\,\dot\psi^2 + \delta_\psi\,\ddot\psi
$$

$$
c_x = 1 + (R'(\delta)\,r_{AO})_x\,\delta_\ell, \qquad c_y = (R'(\delta)\,r_{AO})_y\,\delta_\ell

## Continuous dynamics

Since $a_x$, $a_y$ and $\ddot\delta$ are all linear in $\ddot\ell$, substituting into the Newton--Euler equations gives the 3×3 system:

$$
M(x,t)\begin{bmatrix}\ddot\ell \\ f_{yA} \\ f_{Bn}\end{bmatrix} = \text{rhs}(x,u,t)
$$

Solving for $\ddot\ell$ closes The state-space model:

$$
x = \begin{bmatrix}\ell \\ \dot\ell\end{bmatrix}, \qquad
\dot x = f(x,u,t) = \begin{bmatrix}\dot\ell \\ \ddot\ell\,(x,u,t)\end{bmatrix}$$

---

## RK4 discretization

To simulate or optimize the trajectory, the Continuous model is discretized with fixed step $h$:

$$k_1 = f(x_k,\; u_k,\; t_k)$$

$$k_2 = f\!\left(x_k + \tfrac{h}{2}k_1,\; u_k,\; t_k+\tfrac{h}{2}\right)$$

$$k_3 = f\!\left(x_k + \tfrac{h}{2}k_2,\; u_k,\; t_k+\tfrac{h}{2}\right)$$

$$k_4 = f\!\left(x_k + h\,k_3,\; u_k,\; t_k+h\right)$$

$$x_{k+1} = x_k + \frac{h}{6}\left(k_1 + 2k_2 + 2k_3 + k_4\right)$$

This yields the discrete map $x_{k+1} = F(x_k, u_k)$ used by the trajectory optimizer.

# Modèle dynamique nominal complet en 3 phases

## 1. Hypothèses générales

On considère un bac rigide en mouvement dans le plan, en interaction avec un mur pivotant d’angle imposé $\psi(t)$.

Les hypothèses retenues sont les suivantes :

- le bac est modélisé comme un solide rigide plan ;
- la gravité est négligée ;
- le mur pivotant suit une loi imposée $\psi(t)$ ;
- les forces de réaction de contact ne sont pas calculées explicitement ;
- les contraintes sont prises en compte implicitement via le choix des phases, des coordonnées admissibles et des hypothèses de contact.

On note :

$$
r_{AO}^{body} =
\begin{bmatrix}
a/2 \\
b/2
\end{bmatrix}
\qquad,\qquad
r_{AB}^{body} =
\begin{bmatrix}
0 \\
b
\end{bmatrix}
$$

La matrice de rotation du bac est :

$$
R(\delta)=
\begin{bmatrix}
\cos\delta & -\sin\delta\\
\sin\delta & \cos\delta
\end{bmatrix}
$$

Le centre de masse $O$ a pour moment d’inertie :

$$
I_O=\frac{m}{12}(a^2+b^2)
$$

Le moment d’inertie autour du point $A$ vaut :

$$
I_A = I_O + m\left[\left(\frac a2\right)^2+\left(\frac b2\right)^2\right]
$$

donc :

$$
I_A=\frac{m}{3}(a^2+b^2)
$$

Les efforts extérieurs appliqués au centre de masse sont notés :

$$
F=
\begin{bmatrix}
f_x\\
f_y
\end{bmatrix}
$$

---

## 2. Mur pivotant : géométrie imposée

Le mur pivotant est caractérisé par son angle imposé $\psi(t)$.

Sa direction tangentielle est :

$$
t_B(\psi)=
\begin{bmatrix}
\cos\psi\\
\sin\psi
\end{bmatrix}
$$

Une normale associée au mur est :

$$
n_B(\psi)=
\begin{bmatrix}
\sin\psi\\
-\cos\psi
\end{bmatrix}
$$

La loi du mur est supposée connue :

$$
\psi = \psi(t), \qquad \dot\psi = \dot\psi(t), \qquad \ddot\psi = \ddot\psi(t)
$$

---

## 3. Contraintes de contact

### 3.1. Position du point $B$

Dans le repère monde, la position du point $B$ vaut :

$$
r_B = r_A + R(\delta)\,r_{AB}^{body}
$$

où :

$$
r_{AB}^{body}=
\begin{bmatrix}
0\\
b
\end{bmatrix}
$$

### Phase 1

Comme $A$ est fixe :

$$
r_A=
\begin{bmatrix}
x_A^0\\
y_A^0
\end{bmatrix}
$$

souvent avec $x_A^0=0$, $y_A^0=0$.

Alors :

$$
r_B^{(1)}=
\begin{bmatrix}
x_A^0\\
y_A^0
\end{bmatrix}
+
R(\delta)
\begin{bmatrix}
0\\
b
\end{bmatrix}
$$

d’où :

$$
x_B^{(1)} = x_A^0 - b\sin\delta
$$

$$
y_B^{(1)} = y_A^0 + b\cos\delta
$$

### Phase 2

Comme $A$ glisse sur l’axe des abscisses :

$$
r_A=
\begin{bmatrix}
l\\
0
\end{bmatrix}
$$

donc :

$$
r_B^{(2)}=
\begin{bmatrix}
l\\
0
\end{bmatrix}
+
R(\delta)
\begin{bmatrix}
0\\
b
\end{bmatrix}
$$

d’où :

$$
x_B^{(2)} = l - b\sin\delta
$$

$$
y_B^{(2)} = b\cos\delta
$$

---

### 3.2. Non-pénétration du point $B$ dans le mur

On définit la distance signée du point $B$ au mur par :

$$
\phi_B(q,t)=n_B(\psi(t))^\top r_B(q)
$$

Le contact unilatéral est alors décrit par :

$$
\phi_B(q,t)\ge 0
$$

Interprétation :

- $\phi_B=0$ : contact ;
- $\phi_B>0$ : séparation ;
- $\phi_B<0$ : pénétration interdite.

### Phase 1

$$
\phi_B^{(1)}(\delta,t)=n_B(\psi(t))^\top r_B^{(1)}(\delta)
$$

### Phase 2

$$
\phi_B^{(2)}(l,\delta,t)=n_B(\psi(t))^\top r_B^{(2)}(l,\delta)
$$

$$

\left(\frac a2\cos\delta-\frac b2\sin\delta\right)F_1\cos\psi
+
\left(\frac a2\sin\delta+\frac b2\cos\delta\right)F_1\sin\psi
$$

Donc :

$$
\ddot\delta=
\frac{
\left(\frac a2\cos\delta-\frac b2\sin\delta\right)F_1\cos\psi
+
\left(\frac a2\sin\delta+\frac b2\cos\delta\right)F_1\sin\psi
}{I_A}
$$

### 4.6. Forme d’état

En posant $\omega=\dot\delta$ :

$$
\dot\delta = \omega
$$

$$
\dot\omega = \ddot\delta
$$

### 4.7. Condition de fin de phase

$$
\psi(t)=\psi_{\mathrm{sw},1}
$$

---

## 5. Phase 2 — point $A$ glissant sur l’axe des abscisses

### 5.1. Hypothèses

Pendant la phase 2 :

- le point $A$ peut glisser sur l’axe des abscisses ;
- le point $B$ ne peut pas pénétrer le mur, mais peut s’en détacher ;
- les conditions initiales sont :

$$
l=0,\qquad \dot l=0
$$

- $\delta$ et $\dot\delta$ sont hérités de la phase 1.

### 5.2. Coordonnées généralisées

$$
q=
\begin{bmatrix}
l\\
\delta
\end{bmatrix}
$$

Les états sont :

$$
x_2=
\begin{bmatrix}
l\\
\delta\\
\dot l\\
\dot\delta
\end{bmatrix}
$$

### 5.3. Cinématique du centre de masse

$$
r_A=
\begin{bmatrix}
l\\
0
\end{bmatrix}
$$

$$
r_O^{(2)}=
\begin{bmatrix}
l\\
0
\end{bmatrix}
+
R(\delta)\,r_{AO}^{body}
$$

Donc :

$$
x_O^{(2)}=l+\frac a2\cos\delta-\frac b2\sin\delta
$$

$$
y_O^{(2)}=\frac a2\sin\delta+\frac b2\cos\delta
$$

Les vitesses sont :

$$
\dot x_O^{(2)}
=
\dot l+
\left(-\frac a2\sin\delta-\frac b2\cos\delta\right)\dot\delta
$$

$$
\dot y_O^{(2)}
=
\left(\frac a2\cos\delta-\frac b2\sin\delta\right)\dot\delta
$$

Les accélérations sont :

$$
\ddot x_O^{(2)}
=
\ddot l
+
\left(-\frac a2\sin\delta-\frac b2\cos\delta\right)\ddot\delta
+
\left(-\frac a2\cos\delta+\frac b2\sin\delta\right)\dot\delta^2
$$

$$
\ddot y_O^{(2)}
=
\left(\frac a2\cos\delta-\frac b2\sin\delta\right)\ddot\delta
+
\left(-\frac a2\sin\delta-\frac b2\cos\delta\right)\dot\delta^2
$$

### 5.4. Force de test

On prend une force constante :

$$
f_x^{(2)} = 5.0 \times 0.2 = 1.0
$$

$$
f_y^{(2)} = 5.0 \times (-1.0) = -5.0
$$

### 5.5. Équations dynamiques

Les équations de la phase 2 sont :

$$
m\left[
\ddot l
+
\left(-\frac a2\sin\delta-\frac b2\cos\delta\right)\ddot\delta
+
\left(-\frac a2\cos\delta+\frac b2\sin\delta\right)\dot\delta^2
\right]
= f_x
$$

et

$$
I_A\ddot\delta
+
m\left(-\frac a2\sin\delta-\frac b2\cos\delta\right)\ddot l
=
\left(-\frac a2\sin\delta-\frac b2\cos\delta\right)f_x
+
\left(\frac a2\cos\delta-\frac b2\sin\delta\right)f_y
$$

### 5.6. Forme matricielle

$$
\begin{bmatrix}
m & m\left(-\frac a2\sin\delta-\frac b2\cos\delta\right)\\[4pt]
m\left(-\frac a2\sin\delta-\frac b2\cos\delta\right) & I_A
\end{bmatrix}
\begin{bmatrix}
\ddot l\\
\ddot\delta
\end{bmatrix}
=
\begin{bmatrix}
f_x - m\left(-\frac a2\cos\delta+\frac b2\sin\delta\right)\dot\delta^2\\[4pt]
\left(-\frac a2\sin\delta-\frac b2\cos\delta\right)f_x
+
\left(\frac a2\cos\delta-\frac b2\sin\delta\right)f_y
\end{bmatrix}
$$

Il faut résoudre ce système $2\times2$ pour obtenir $\ddot l$ et $\ddot\delta$.

### 5.7. Forme d’état

En posant :

$$
v_l=\dot l
\qquad,\qquad
\omega=\dot\delta
$$

on obtient :

$$
\dot l = v_l
$$

$$
\dot\delta = \omega
$$

$$
\dot v_l = \ddot l
$$

$$
\dot\omega = \ddot\delta
$$

### 5.8. Conditions initiales

Au début de la phase 2 :

$$
l=0,\qquad \dot l=0
$$

$$
\delta=\delta_{\mathrm{fin\,phase\,1}}
\qquad,\qquad
\dot\delta=\dot\delta_{\mathrm{fin\,phase\,1}}
$$

### 5.9. Condition de fin de phase

La phase 2 se termine lorsqu’un nouveau point du bac entre en contact.

## 6. Transition phase 2 $\rightarrow$ phase 3 : impact sans rebond

### 6.1. Hypothèse d’impact

On suppose un impact sans rebond :

- les positions restent continues ;
- les vitesses changent instantanément ;
- il n’y a pas de restitution ;
- l’énergie cinétique peut diminuer.

On note $q^-$ et $\dot q^-$ les grandeurs juste avant impact, et $q^+$, $\dot q^+$ juste après impact.

Les positions vérifient :

$$
q^+ = q^-
$$

Les vitesses vérifient une projection sur le nouvel espace admissible :

$$
\dot q^+ = \Pi(q^-)\,\dot q^-
$$

où $\Pi(q^-)$ représente une projection imposant la nouvelle contrainte de contact.

---

## 7. Phase 3 — nouveau contact maintenu

### 7.1. Hypothèses

Pendant la phase 3 :

- un nouveau point du bac est en contact ;
- ce contact est supposé maintenu pendant toute la phase ;
- la phase commence après un impact sans rebond ;
- les vitesses initiales de phase 3 sont donc les vitesses projetées après impact.

### 7.2. Forme générale de la contrainte

On note la contrainte géométrique active :

$$
\phi_3(q)=0
$$

Sa jacobienne est :

$$
J_3(q)=\frac{\partial \phi_3}{\partial q}
$$

### 7.3. Dynamique contrainte

La dynamique générale s’écrit :

$$
M(q)\ddot q + J_3(q)^T\lambda = b(q,\dot q,u)
$$

avec la contrainte d’accélération :

$$
J_3(q)\ddot q + \dot J_3(q,\dot q)\dot q = 0
$$

où :

- $M(q)$ est la matrice d’inertie ;
- $b(q,\dot q,u)$ regroupe les efforts extérieurs et termes géométriques ;
- $\lambda$ représente la réaction associée à la contrainte.

### 7.4. Si l’on veut éviter $\lambda$

Si l’on ne veut toujours pas calculer explicitement les réactions, il faut reformuler la phase 3 dans un espace admissible réduit, c’est-à-dire choisir une nouvelle coordonnée généralisée adaptée à la contrainte active.

## 9. Résumé global du système

### Phase 1

États :

$$
x_1=
\begin{bmatrix}
\delta\\
\dot\delta
\end{bmatrix}
$$

Équation :

$$
I_A\ddot\delta
=
\left(\frac a2\cos\delta-\frac b2\sin\delta\right)f_y
-
\left(\frac a2\sin\delta+\frac b2\cos\delta\right)f_x
$$

Contrainte :

$$
\phi_B^{(1)}(\delta,t)\ge 0
$$

---

### Phase 2

États :

$$
x_2=
\begin{bmatrix}
l\\
\delta\\
\dot l\\
\dot\delta
\end{bmatrix}
$$

Équations :

$$
m\left[
\ddot l
+
\left(-\frac a2\sin\delta-\frac b2\cos\delta\right)\ddot\delta
+
\left(-\frac a2\cos\delta+\frac b2\sin\delta\right)\dot\delta^2
\right]
= f_x
$$

$$
I_A\ddot\delta
+
m\left(-\frac a2\sin\delta-\frac b2\cos\delta\right)\ddot l
=
\left(-\frac a2\sin\delta-\frac b2\cos\delta\right)f_x
+
\left(\frac a2\cos\delta-\frac b2\sin\delta\right)f_y
$$

Contrainte :

$$
\phi_B^{(2)}(l,\delta,t)\ge 0
$$

---

### Phase 3

Transition :

$$
q^+=q^-
$$

$$
\dot q^+ = \Pi(q^-)\dot q^-
$$

Phase continue :

$$
M(q)\ddot q + J_3(q)^T\lambda = b(q,\dot q,u)
$$

avec :

$$
\phi_3(q)=0
$$

## 10. Interprétation

Le modèle ainsi obtenu est un **modèle dynamique nominal multi-phases**. Les contraintes ne sont pas absentes du système ; elles sont intégrées :

- en phase 1 par l’hypothèse d’un point $A$ fixe ;
- en phase 2 par l’hypothèse d’un point $A$ glissant sur l’axe des abscisses ;
- en phases 1 et 2 par la non-pénétration du point $B$ vis-à-vis du mur ;
- en phase 3 par un impact sans rebond suivi d’un contact maintenu.

Ce modèle sert à décrire un mouvement nominal cohérent du bac, pouvant ensuite être utilisé comme base de référence pour une stratégie de commande.
